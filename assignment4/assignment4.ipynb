```python
import numpy as np
import random
import matplotlib.pyplot as plt
import scipy
import scipy.stats
%matplotlib inline
```

```python
mean = 700
t = 4000
n = 10000
X_t = (1/mean)*np.exp(-(1/mean)*t)
```

```python
X = np.random.exponential(mean, n)
X_m = [x for x in X if x>X_t]        
```

```python
def MLE_exp(X_m,n,t):
    m = len(X_m)
    sum_X_m = sum(X_m)
    return (sum_X_m+(n-m)*t)/m
```

```python
est_mean = MLE_exp(X_m,n,t)
print(f"Estimated Mean:{est_mean}\nTrue Mean:{mean}")
```

```python
X_est = np.random.exponential(est_mean, n)
```

```python
_, l_est = scipy.stats.expon.fit(X_est, floc=0)
_, l = scipy.stats.expon.fit(X, floc=0)
```

```python
pdf = scipy.stats.expon.pdf
x = range(0, 5000)
plt.figure(figsize=(15,8))
plt.hist(X, bins=x, density='true')
plt.plot(pdf(x, scale=l),color='b')
plt.plot(pdf(x, scale=l_est),color='r',linestyle='dotted')
plt.xlabel('Hours')
plt.ylabel('Observed PDF')
plt.legend(['Fitted Distribution on True Mean', 'Fitted Distribution on Estimate Mean', ]);
```
